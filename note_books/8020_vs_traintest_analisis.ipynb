### <font color=#785964> Train & test files vs Train splited 80/20:</font>  

```python
data_of_8020 = test_vs_8020[test_vs_8020['split80/20']=='yes']
data_of_train_n_test = test_vs_8020[test_vs_8020['split80/20']=='no']

# Accuracy = TP+TN/TP+FP+FN+TN
Accuracy_8020 = (data_of_8020['TP'] + data_of_8020['TN']) /\
        (data_of_8020['TP'] + data_of_8020['TN'] + data_of_8020['FP'] + data_of_8020['FN'])

Accuracy_test = (data_of_train_n_test['TP'] + data_of_train_n_test['TN'] )/\
        (data_of_train_n_test['TP'] + data_of_train_n_test['TN'] + data_of_train_n_test['FP'] + data_of_train_n_test['FN'])

# Precision = TP/TP+FP
Precision_8020 = data_of_8020['TP'] / (data_of_8020['TP'] + data_of_8020['FP'])
Precision_test = data_of_train_n_test['TP'] /( data_of_train_n_test['TP'] + data_of_train_n_test['FP'])

# Recall = TP/TP+FN
Recall_8020 = data_of_8020['TP'] / (data_of_8020['TP'] + data_of_8020['FN'])
Recall_test = data_of_train_n_test['TP'] / (data_of_train_n_test['TP'] + data_of_train_n_test['FN'])

# F1_Score = 2*(Recall * Precision) / (Recall + Precision)
F1_score_8020 = 2 * (Recall_8020 * Precision_8020) / (Recall_8020 + Precision_8020)
F1_score_test = 2 * (Recall_test * Precision_test) / (Recall_test + Precision_test)

test_vs_8020.loc[test_vs_8020['split80/20'] == 'yes', 'Accuracy'] = Accuracy_8020
test_vs_8020.loc[test_vs_8020['split80/20'] == 'no', 'Accuracy'] = Accuracy_test
test_vs_8020.loc[test_vs_8020['split80/20'] == 'yes', 'Precision'] = Precision_8020
test_vs_8020.loc[test_vs_8020['split80/20'] == 'no', 'Precision'] = Precision_test
test_vs_8020.loc[test_vs_8020['split80/20'] == 'yes', 'Recall'] = Recall_8020
test_vs_8020.loc[test_vs_8020['split80/20'] == 'no', 'Recall'] = Recall_test
test_vs_8020.loc[test_vs_8020['split80/20'] == 'yes', 'F1_Score'] = F1_score_8020
test_vs_8020.loc[test_vs_8020['split80/20'] == 'no', 'F1_Score'] = F1_score_test


# ploting matrix area  stackbar for train/test start 
# Values of each group
columns = ['TP','FP','TN','FN']
values = test_vs_8020[test_vs_8020['split80/20']=='no'][columns]
bars1 = values['TP'].to_list()
bars2 = values['FN'].to_list()
bars3 = values['TN'].to_list()
bars4 = values['FP'].to_list()

# Heights of bars1 + bars2
bars_h1 = np.add(bars1, bars2).tolist()
bars_h2 = np.add(bars_h1, bars3).tolist()
 
# The position of the bars on the x-axis
r = [0,1,2,3,4,5]
 
# Names of group and bar width
names = test_vs_8020[test_vs_8020['split80/20']=='no']['function'].to_list()
barWidth = 1
 
# Create brown bars
plt.bar(r, bars1, color='#fb5607', edgecolor='white', width=barWidth, label='TP')
# Create green bars (middle1), on top of the second ones
plt.bar(r, bars2, bottom=bars1, color='#3a86ff', edgecolor='white', width=barWidth, label='FN')
# Create green bars (middle2), on top of the two first
plt.bar(r, bars3, bottom=bars_h1, color='#ff006e', edgecolor='white', width=barWidth, label='TN')
# Create green bars (top)
plt.bar(r, bars4, bottom=bars_h2, color='#8338ec', edgecolor='white', width=barWidth, label='FP')
 
# Custom X axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel("train - text")
 
# Show graphic
fig.tight_layout()
plt.legend()
plt.show()
# ploting matrix area  stackbar for train/test end 

# ploting matrix area  stackbar for 8020 start 
columns = ['TP','FP','TN','FN']
values = test_vs_8020[test_vs_8020['split80/20']=='yes'][columns]
bars1 = values['TP'].to_list()
bars2 = values['FN'].to_list()
bars3 = values['TN'].to_list()
bars4 = values['FP'].to_list()

# Heights of bars1 + bars2
bars_h1 = np.add(bars1, bars2).tolist()
bars_h2 = np.add(bars_h1, bars3).tolist()
 
# The position of the bars on the x-axis
r = [0,1,2,3,4,5]
 
# Names of group and bar width
names = test_vs_8020[test_vs_8020['split80/20']=='yes']['function'].to_list()
barWidth = 1
 
# Create brown bars
plt.bar(r, bars1, color='#fb5607', edgecolor='white', width=barWidth, label='TP')
# Create green bars (middle1), on top of the second ones
plt.bar(r, bars2, bottom=bars1, color='#3a86ff', edgecolor='white', width=barWidth, label='FN')
# Create green bars (middle2), on top of the two first
plt.bar(r, bars3, bottom=bars_h1, color='#ff006e', edgecolor='white', width=barWidth, label='TN')
# Create green bars (top)
plt.bar(r, bars4, bottom=bars_h2, color='#8338ec', edgecolor='white', width=barWidth, label='FP')
 
# Custom X axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel("80/20")

# Show graphic
fig.tight_layout()
plt.legend()
plt.show()
# ploting matrix area  stackbar for 8020 end 
 


# ploting scores:
fig, axes = plt.subplots(1,4,figsize=(20, 5)) 
mylabels = ('80/20','test')
for i,col in enumerate(['Accuracy','Precision','Recall','F1_Score']):
    sns.barplot(x="function", y=col, hue="split80/20", data=test_vs_8020 , ax = axes[i])  
    handle, labels = axes[i].get_legend_handles_labels()
    axes[i].legend(handle, mylabels)
fig.tight_layout()
plt.show()
```
